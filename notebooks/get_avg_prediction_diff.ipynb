```python
import pandas as pd

df = pd.read_csv("/home/hjall/work/qai/xai/xai-nlp-benchmark/artifacts/xai-nlp-benchmark-2024-04-23-21-20-02/visualization_gender_no_sub_samp/prediction_diff.csv")
df.rename(columns={
    "mean": "mean_mean",
    "std": "std_mean",
})
```

```python
from visualization.common import MODEL_ORDER

df_all = df[df["dataset"] == "gender_all"]
df_subj = df[df["dataset"] == "gender_subj"]

def print_table_row(cur_df: pd.DataFrame):
    print(cur_df["dataset"].iloc[0])
    ret = "Avg. Prediction Diff."
    for model in MODEL_ORDER:
        cur_model = cur_df[cur_df["model"] == model]
        mean = cur_model["mean"].iloc[0] * 100
        std = cur_model["mean_std"].iloc[0] * 100

        ret += f" & {mean:.2f} $\pm$ {std:.2f}"
    print(ret + " \\\\")

print_table_row(df_all)
print_table_row(df_subj)
```
