```python
from visualization.main import load_xai_records
from utils import load_jsonl_as_df

gender_all = "/home/hjall/work/qai/xai/xai-nlp-benchmark/artifacts/data/nlp-benchmark_2024-04-14-16-58-32/gender_all/train.jsonl"
gender_subj = "/home/hjall/work/qai/xai/xai-nlp-benchmark/artifacts/data/nlp-benchmark_2024-04-14-16-58-32/gender_subj/train.jsonl"

df_all = load_jsonl_as_df(gender_all)
df_all["type"] = "all"
df_subj = load_jsonl_as_df(gender_subj)
df_subj["type"] = "subj"

df = df_all.append(df_subj)
df["length"] = df["sentence"].apply(lambda x: len(x))
```

```python
df = df[df["length"] < 12]
df
```

```python
import pandas as pd

def get_gt_words(group_df: pd.DataFrame) -> list[str]:
    gt_words = []
    for _, row in group_df.iterrows():
        for idx, word in enumerate(row["sentence"]):
            if row["ground_truth"][idx] == 1:
                gt_words.append(word)
    return gt_words

for sentence_idx, group in df.groupby("sentence_idx"):
    group_all = group[group["type"] == "all"]
    group_subj = group[group["type"] == "subj"]

    gt_words_all = get_gt_words(group_all) 
    gt_words_subj = get_gt_words(group_subj)
    print(f"Sentence_idx: {sentence_idx}")
    print(gt_words_all)
    print(gt_words_subj)
    print("")
    
    
```

```python
rows = df[df["sentence_idx"] == 1241]

fem_all = rows[(rows["type"] == "all") & (rows["target"] == 0)].iloc[0]
fem_subj = rows[(rows["type"] == "subj") & (rows["target"] == 0)].iloc[0]

male_all = rows[(rows["type"] == "all") & (rows["target"] == 1)].iloc[0]
male_subj = rows[(rows["type"] == "subj") & (rows["target"] == 1)].iloc[0]

def print_line(row, base):
    ret = base
    for idx, word in enumerate(row["sentence"]):
        if row["ground_truth"][idx] == 1:
            ret += f"\\textbf{{{word}}}"
        else:
            ret += word
        if idx != len(row["sentence"]) - 1:
            ret += " & "
        else:
            ret += " "
    
    print(ret + " \\\\")

print_line(fem_subj, "Subj Female &")
print_line(male_subj, "Subj Male &")
print_line(fem_all, "All Female &")
print_line(male_all, "All Male &")


```
